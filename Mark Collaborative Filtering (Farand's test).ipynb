```python
product_matrix=data.pivot_table(index="user_id",columns='product_id',values="interaction",fill_value=0)
product_matrix
```

```python
product_matrix.shape
```

```python
product_matrix.query('user_id ==512367874').sum(axis=1)
```

```python
product_matrix=product_matrix.T
product_matrix
```

```python
product_matrix.shape
```

```python
import sklearn
from sklearn.decomposition import TruncatedSVD
SVD = TruncatedSVD(n_components=10)
decomposed_matrix = SVD.fit_transform(product_matrix)
decomposed_matrix.shape
```

```python
correlation_matrix = np.corrcoef(decomposed_matrix)
correlation_matrix.shape
```
